## First Objective: How does start date affect length to approval?

```python
permit_df['start_month']=''
for j in range(0,permit_df_shape[0]):
    permit_df['start_month'].iloc[j]=permit_df['application_start_date'].iloc[j][5:7]
permit_df['start_month'] = pd.to_numeric(permit_df['start_month'])
```

```python
permit_df['start_month']
```

```python
month_proc_time_gb=permit_df[['start_month','processing_time']].groupby(['start_month'], as_index=False)

month_proc_time_df=pd.DataFrame(month_proc_time_gb.mean())

month_proc_time_df
```

```python
month_num_applications_df=pd.DataFrame(month_proc_time_gb.count())

month_num_applications_df
```

## Objective 3 How do permits per population in a zip code compare for diffenet zip codes.  

```python
# Data from https://datahub.cmap.illinois.gov/dataset/2010-census-data-summarized-to-chicago-community-areas

# File to Load (Remember to Change These)
file_to_load = "data/CCASF12010CMAP.csv"

# Read Purchasing File and store into Pandas dataframe
population_data = pd.read_csv(file_to_load)
population_data.head()
```

```python
zip_ct_df=pd.DataFrame(zip_proc_time_gb.count())

zip_ct_df['Total Population']=''
for j in range(0,zip_ct_df.shape[0]):
    zip_ct_df['Total Population'].iloc[j]=population_data['P0050001'].iloc[int(zip_ct_df['community_area'].iloc[j])]
zip_ct_df['Total Population']=pd.to_numeric(zip_ct_df['Total Population'])

zip_ct_df['perm_p_person']=(zip_ct_df['processing_time']/zip_ct_df['Total Population'])
    
zip_ct_df
```
