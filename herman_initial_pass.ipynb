```python
import pandas as pd
```

```python
permit_df = pd.read_json("https://data.cityofchicago.org/resource/building-permits.json")
```

```python
permit_df.head()
```

```python
permit_df_shape=permit_df.shape

permit_df_shape[0]
```

## First Objective: How does start date affect length to approval?

```python
permit_df['start_month']=''
for j in range(0,permit_df_shape[0]):
    permit_df['start_month'].iloc[j]=permit_df['application_start_date'].iloc[j][5:7]
permit_df['start_month'] = pd.to_numeric(permit_df['start_month'])
```

```python
permit_df['start_month']
```

```python
month_proc_time_gb=permit_df[['start_month','processing_time']].groupby(['start_month'], as_index=False)

month_proc_time_df=pd.DataFrame(month_proc_time_gb.mean())

month_proc_time_df
```

## Second Objective: How does zip code affect length to approval?

```python
zip_proc_time_gb=permit_df[['street_zip','processing_time']].groupby(['street_zip'], as_index=False)

zip_proc_time_df=pd.DataFrame(month_proc_time_gb.mean())

zip_proc_time_df
```
